```python
def catogery(df, col, labels):
    desc = df[col].describe()
    edges = [desc['min'], desc['25%'], desc['50%'], desc['75%'], desc['max']]
    df[col + '_category'] = pd.cut(df[col], edges, labels=labels, duplicates='drop')
    return df
df['Vote_Average'] = pd.to_numeric(df['Vote_Average'], errors='coerce')

  
```

```python
labels = ['not popular', 'below_avg', 'avg', 'popular']
df = catogery(df, 'Vote_Average', labels)
print(df[['Vote_Average', 'Vote_Average_category']].head())

```
